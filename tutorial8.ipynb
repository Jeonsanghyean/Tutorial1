## Functions in Python

```python
def myfun(x, y):
    return x * y
```

```python
myfun(3, 4)
```

```python
def myfun(x):
    return x*3
```

```python
def apply(q, x):
    return q(x)
```

```python
apply(myfun, 7)
```

```python
apply(lambda z: z * 4, 7)
```
