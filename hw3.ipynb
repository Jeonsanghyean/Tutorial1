```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("IMG_6504.jpg")
row,cols,ch = img.shape
res1 = cv2.resize(img,None,fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
res2 = cv2.resize(img,None,fx=2, fy=2, interpolation=cv2.INTER_AREA)
res3 = cv2.resize(img,None,fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
```

```python
cv2.namedWindow('res1')
cv2.namedWindow('res2')
cv2.namedWindow('res3')
cv2.resizeWindow('res1', row*2, cols*2)
cv2.resizeWindow('res2', row*2, cols*2)
cv2.resizeWindow('res3', row*2, cols*2)
```

```python
while(1):
    cv2.imshow('res1',res1)
    cv2.imshow('res2',res2)
    cv2.imshow('res3',res3)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```

```python
img = cv2.imread('IMG_6504.jpg')
pts1 = np.float32([[171,171],[764,171],[805,1100],[85,1100]])
pts2 = np.float32([[0,0],[900,0],[900,1200],[0,1200]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(900,1200))
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
