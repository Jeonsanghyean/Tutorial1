```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture('trk1.mp4')
```

```python
while(cap.isOpened()):
  ret, frame = cap.read()
  hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
  lower_orange = np.array([10, 100, 100])  # 하한값 (색상, 채도, 명도)
  upper_orange = np.array([25, 255, 255])  # 상한값 (색상, 채도, 명도)
  mask = cv2.inRange(hsv, lower_orange, upper_orange)
  res = cv2.bitwise_and(frame, frame, mask=mask)
  cv2.imshow('frame', frame)
  cv2.imshow('mask', mask)
  cv2.imshow('res', res)

  k = cv2.waitKey(5) & 0xFF
  if k == 27:
    break
  
cv2.destroyAllWindows()
```
